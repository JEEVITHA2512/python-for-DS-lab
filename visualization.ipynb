```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
iris=load_iris()
x=iris.data
y=iris.target
```

```python
#histogram
plt.hist(x[:,0],bins=30)
plt.xlabel('sepal length')
plt.ylabel('frequency')
plt.title('histogram of sepal length')
plt.show()
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.title('scatterplot of sepal length vs sepal width')
plt.show()
```

```python
plt.plot(x[:,2])
plt.xlabel('index')
plt.ylabel('petal length')
plt.title('line plot of petal length')
plt.show()
```

```python
a=[x[y==i,0].mean()for i in range(3)]
target_names=iris.target_names
plt.bar(target_names,a)
plt.xlabel('target class')
plt.ylabel('average sepal length')
plt.title('bar plot of average sepal length for each target class')
```

```python
proportions=[sum(y==i)for i in range(3)]
plt.pie(proportions,labels=target_names)
plt.title('pie chart ')
plt.show()
```
